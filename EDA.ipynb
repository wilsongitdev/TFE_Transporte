##  Importar librerias

```python
import csv
import pandas as pd
```

## Leyendo datos masivos

```python
chunksize = 500000
all_chunks = []
for chunk in pd.read_csv('/kaggle/input/datos-trafico/datos_procesados.csv', sep=";", decimal=".", chunksize=chunksize):
  all_chunks.append(chunk)

df = pd.concat(all_chunks, ignore_index=True)
df_eda = df.iloc[:, 2:]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = df_eda.select_dtypes(exclude=['object']).columns

for col in columns:
    plt.figure(figsize=(5, 7))
    sns.boxplot(data=df_eda[col])
    plt.title(f"Diagrama de Caja de {col}")
    plt.ylabel('Valores')
    plt.show()
```

###  Análisis Univariado de características (variables numéricas)

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_var_col = df_eda.select_dtypes(include=['float64', 'int64']).columns
for col in cat_var_col:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_eda[col], kde=True)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

###  Análisis Univariado de características (variable categoricas)

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cat_col = df_eda.select_dtypes(include=['object']).columns
for col in num_cat_col:
    plt.figure(figsize=(15, 10))
    sns.histplot(data=df_eda[col])
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

### Análisis Bivariado de características

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Análisis Bivariado
# Relación entre vmed e intensidad
sns.scatterplot(x='vmed', y='intensidad', data=df_eda)
plt.title('Relación entre v_med y intensidad')
plt.xlabel('v_med')
plt.ylabel('intensidad')
plt.show()

# Relación entre v_med y carga
sns.scatterplot(x='vmed', y='carga', data=df_eda)
plt.title('Relación entre v_med y carga')
plt.xlabel('v_med')
plt.ylabel('carga')
plt.show()

# Relación entre intensidad y carga
sns.scatterplot(x='intensidad', y='carga', data=df_eda)
plt.title('Relación entre intensidad y carga')
plt.xlabel('intensidad')
plt.ylabel('carga')
plt.show()

# Relación entre temperatura y carga
sns.scatterplot(x='TEMPERATURA', y='intensidad', data=df_eda)
plt.title('Relación entre temperatura y intensidad')
plt.xlabel('temperatura')
plt.ylabel('intensidad')
plt.show()

# Relación entre precipitacion e intensidad
sns.scatterplot(x='PRECIPITACION', y='intensidad', data=df_eda)
plt.title('Relación entre precipitación y intensidad')
plt.xlabel('precipitación')
plt.ylabel('intensidad')
plt.show()


# Relación entre festivo y intensidad
sns.boxplot(x='FESTIVO', y='intensidad', data=df_eda)
plt.title('Relación entre festivo y intensidad')
plt.xlabel('festivo')
plt.ylabel('intensidad')
plt.show()
```

```python

numericData = df_eda.loc[:,['intensidad','ocupacion' , 'carga', 'vmed','ANO','MES', 'DIA', 'HORA', 'MINUTO', 'DIA_SEMANA','TEMPERATURA','PRECIPITACION','FESTIVO']]

# Calcular la matriz de correlación
correlationMatrixPearson = numericData.corr()

# Mostrar la matriz de correlación
plt.figure(figsize=(10, 10))
sns.heatmap(correlationMatrixPearson, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```
