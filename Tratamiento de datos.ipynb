### Selección de las estaciones  
Vamos a seleccionar estaciones que no tengan un gran número de datos faltantes.

```python
# Crear el DataFrame con la cuenta de registros por cada valor de "id"
conteo_por_id_2023 = datos2023.groupby('id').size().reset_index(name='cantidad')
conteo_por_id_2024 = datos2024.groupby('id').size().reset_index(name='cantidad')

# Encontrar la frecuencia máxima
frecuencia_maxima_2023 = conteo_por_id_2023['cantidad'].max()
frecuencia_maxima_2024 = conteo_por_id_2024['cantidad'].max()

# Filtrar el DataFrame para quedarse solo con los valores cuya frecuencia es la máxima
conteo_maximos_2023 = conteo_por_id_2023[conteo_por_id_2023['cantidad'] >= frecuencia_maxima_2023-15]
conteo_maximos_2024 = conteo_por_id_2024[conteo_por_id_2024['cantidad'] >= frecuencia_maxima_2024-15]

# Filtrar el DataFrame original "datos" para que solo incluya los registros cuyo "id" esté en "conteo_maximos"
datos_filtrados_2023 = datos2023[datos2023['id'].isin(conteo_maximos_2023['id'])]
datos_filtrados_2024 = datos2024[datos2024['id'].isin(conteo_maximos_2024['id'])]
```

Tras analizar las estaciones con mejores datos nos quedamos con las siguientes (todas pertenecientes al entorno de la M-30)

```python
# Lista de valores de "id" proporcionados
ids_filtrar = [
    6642,6735,6667,6783,7127,3703,6744,3539,6661,10176,3536,6663,6731,6793,10661,3820,6682,6693,6680,6755,6753,6807,6738,10179,6786,10178,
    6728,3531,6819,6756,6778,3705,3800,7013,6686,6777,6668,6669,10267,6742,6741,6739,6684,6822,3490,10268,6674,6750,6820,3838,6758,6676,
    7012,6719,3817    
]


# Filtrar ambos DataFrames
datos2023_filtrados = datos2023[datos2023['id'].isin(ids_filtrar)]
datos2024_filtrados = datos2024[datos2024['id'].isin(ids_filtrar)]

# Unión de datos2023_filtrados y datos2024_filtrados
union_datos = pd.concat([datos2023_filtrados, datos2024_filtrados], ignore_index=True)
```

### Tratamiento de los datos  
A continuación se realizan unas transformaciones previas que dejarás los datos listos para el EDA

```python
# Convertimos el formato de la columna fecha
union_datos['fecha'] = pd.to_datetime(union_datos['fecha'])
```

```python
# Dividimos la fecha en varias columnas
union_datos['ANO'] = union_datos['fecha'].dt.year
union_datos['MES'] = union_datos['fecha'].dt.month
union_datos['DIA'] = union_datos['fecha'].dt.day
union_datos['HORA'] = union_datos['fecha'].dt.hour
union_datos['MINUTO'] = union_datos['fecha'].dt.minute
```

```python
# Mostramos los datos
union_datos.head()
```

```python
# Creamos un dataframe con todas las fechas posibles
todas_fechas = union_datos[['ANO', 'MES', 'DIA', 'HORA', 'MINUTO']].drop_duplicates()
```

 A continuación vamos a generar los registros faltantes. Si a una estación le falta algún registro se le va a crear con valores nulos.

```python
# Agregar la columna 'id' a todas_fechas cruzando todos los posibles id de datos
id_fechas = pd.merge(
    union_datos[['id']].drop_duplicates(),  # Tomar ids únicos de 'datos'
    todas_fechas,
    how='cross'  # Generar todas las combinaciones de 'id' y 'todas_fechas'
)

# Realizar el merge para identificar qué registros de id_fechas no tienen coincidencias en datos
merge_result = pd.merge(
    id_fechas,
    union_datos,
    on=['id', 'ANO', 'MES', 'DIA', 'HORA', 'MINUTO'],
    how='left',
    indicator=True  # Agregar columna indicador
)

# Filtrar registros de id_fechas que no tienen coincidencias en datos
sin_coincidencia = merge_result[merge_result['_merge'] == 'left_only']

# Crear un DataFrame con los registros faltantes en `datos` pero con columnas nulas
nuevos_registros = sin_coincidencia[['id', 'ANO', 'MES', 'DIA', 'HORA', 'MINUTO']].copy()

# Añadir las columnas de datos con valores nulos
nuevos_registros['fecha'] = np.nan
nuevos_registros['tipo_elem'] = np.nan
nuevos_registros['intensidad'] = np.nan
nuevos_registros['ocupacion'] = np.nan
nuevos_registros['carga'] = np.nan
nuevos_registros['vmed'] = np.nan
nuevos_registros['error'] = np.nan
nuevos_registros['periodo_integracion'] = np.nan

# Concatenar estos nuevos registros con el DataFrame original 'datos'
resultado_final = pd.concat([union_datos, nuevos_registros], ignore_index=True)

# Rellenamos la columna error, si el valor es nulo es porque lo acabamos de rellenar nosotros y es claramente un error
resultado_final['error'] = resultado_final['error'].fillna('S')

# Mostrar el DataFrame final
resultado_final.head()
```

```python
# Comprobamos que todos los identificadores tienen el mismo número de valores
resultado_final['id'].value_counts(ascending=False)
```

```python
# Queremos añadir el día de la semana
# Lo primero es reconstruir la columna fecha, ya que para los registros nulos tendrá valor null
resultado_final['fecha'] = pd.to_datetime(
    resultado_final.rename(columns={'ANO': 'year', 'MES': 'month', 'DIA': 'day', 'HORA': 'hour', 'MINUTO': 'minute'})[['year', 'month', 'day', 'hour', 'minute']]
)

# Convertir a tipo date
resultado_final['fecha'] = pd.to_datetime(resultado_final['fecha'])

resultado_final.head()
```

```python
# Creamos una nueva columna con el día de la semana
resultado_final['DIA_SEMANA'] = resultado_final['fecha'].dt.weekday.astype(int)

resultado_final.head()
```

Una vez hechos los tratamientos de datos previos, guardamos el resultado en ficheros csv

```python
resultado_final.to_csv('datos_finales.csv', index=False, sep=';')
```

# Datos meteorológicos

### Lectura de los datos

```python
# Cargamos los datos meteorológicos
datos012023 = pd.read_csv('meteo0123.csv', sep=';', decimal='.')
datos022023 = pd.read_csv('meteo0223.csv', sep=';', decimal='.')
datos032023 = pd.read_csv('meteo0323.csv', sep=';', decimal='.')
datos042023 = pd.read_csv('meteo0423.csv', sep=';', decimal='.')
datos052023 = pd.read_csv('meteo0523.csv', sep=';', decimal='.')
datos062023 = pd.read_csv('meteo0623.csv', sep=';', decimal='.')
datos072023 = pd.read_csv('meteo0723.csv', sep=';', decimal='.')
datos082023 = pd.read_csv('meteo0823.csv', sep=';', decimal='.')
datos092023 = pd.read_csv('meteo0923.csv', sep=';', decimal='.')
datos102023 = pd.read_csv('meteo1023.csv', sep=';', decimal='.')
datos112023 = pd.read_csv('meteo1123.csv', sep=';', decimal='.')
datos122023 = pd.read_csv('meteo1223.csv', sep=';', decimal='.')

datos012024 = pd.read_csv('meteo0124.csv', sep=';', decimal='.')
datos022024 = pd.read_csv('meteo0224.csv', sep=';', decimal='.')
datos032024 = pd.read_csv('meteo0324.csv', sep=';', decimal='.')
datos042024 = pd.read_csv('meteo0424.csv', sep=';', decimal='.')
datos052024 = pd.read_csv('meteo0524.csv', sep=';', decimal='.')
datos062024 = pd.read_csv('meteo0624.csv', sep=';', decimal='.')
datos072024 = pd.read_csv('meteo0724.csv', sep=';', decimal='.')
datos082024 = pd.read_csv('meteo0824.csv', sep=';', decimal='.')
datos092024 = pd.read_csv('meteo0924.csv', sep=';', decimal='.')
datos102024 = pd.read_csv('meteo1024.csv', sep=';', decimal='.')

# Concatenamos los meses
datos2023 = pd.concat([datos012023,datos022023,datos032023,datos042023,datos052023,datos062023,datos072023,datos082023,datos092023,datos102023,datos112023,datos122023])
datos2024 = pd.concat([datos012024,datos022024,datos032024,datos042024,datos052024,datos062024,datos072024,datos082024,datos092024,datos102024])

datos_meteo = pd.concat([datos2023,datos2024])

# Mostramos los primeros registros
datos_meteo.head()
```

### Selección de magnitudes y estación  
Entre todos los parámetros que mide la estación, vamos a quedarnos con la temperatura y la precipitación.  
Al ser todas las estaciones de la ciudad de Madrid, nos vamos a quedar con aquella que tenga los mejores datos.

```python
# Seleccionamos la temperatura y la precipitación
datos_meteo = datos_meteo[datos_meteo.MAGNITUD.isin([83,89])]

# Vemos el número de registros de las estaciones
datos_meteo.groupby('ESTACION')['MAGNITUD'].count()
```

```python
# Nos quedamos con la estación 54
datos_meteo = datos_meteo[datos_meteo.ESTACION.isin([54])]
# Nos quedamos con las magnitudes 83 (Temperatura) y 89 (Precipitación)
datos_meteo = datos_meteo[datos_meteo.MAGNITUD.isin([83,89])]

# Eliminamos las columnas PROVINCIA y MUNICIPIO ya que todas las mediciones se corresponden con Madrid
datos_meteo = datos_meteo.drop(columns=['PROVINCIA', 'MUNICIPIO'])

# Eliminamos las columnas de validación.
datos_meteo = datos_meteo.loc[:, ~datos_meteo.columns.str.startswith('V')]

# Mostramos los resultados
datos_meteo
```

### Tratamiento de los datos  
Al igual que se hizo con los datos del tráfico, vamos a realizar un tratamiento previo al EDA.  
El primer paso es covertir los datos en un registro por día a un registro por hora.

```python
# Definir las columnas H (H01, H02, ..., H24)
columnas_h = [f'H{i:02d}' for i in range(1, 25)]

# Aplicar melt para convertir las columnas H en filas
datos_meteo_long = datos_meteo.melt(id_vars=['ESTACION', 'MAGNITUD', 'PUNTO_MUESTREO', 'ANO', 'MES', 'DIA'],
                                    value_vars=columnas_h,
                                    var_name='HORA',
                                    value_name='VALOR')

# Reemplazar los valores de la columna 'Hora' para que H01 se convierta en 1, H02 en 2, etc.
datos_meteo_long['HORA'] = datos_meteo_long['HORA'].str.replace('H', '').astype(int)

# Mostrar el resultado
datos_meteo_long.head()
```

```python
#Eliminamos la H y convertimos el resultado a entero
#datos_meteo_long['HORA'] = datos_meteo_long['HORA'].str.replace(r'H(\d{2})', r'\1').astype(int)
# Restamos 1 al valor de la columna Hora
datos_meteo_long['HORA'] = datos_meteo_long['HORA'] - 1
datos_meteo_long
```

Una vez hechos los tratamientos de datos previos, guardamos el resultado en ficheros csv

```python
# Crear un dataset con los registros donde MAGNITUD es 83
datos_meteo_83 = datos_meteo_long[datos_meteo_long['MAGNITUD'] == 83]

# Crear un dataset con los registros donde MAGNITUD es 89
datos_meteo_86 = datos_meteo_long[datos_meteo_long['MAGNITUD'] == 89]
```

```python
datos_meteo_83.to_csv('datos_meteo_83.csv', index=False, sep=';')
datos_meteo_86.to_csv('datos_meteo_89.csv', index=False, sep=';')
```

# Datos de los días festivos  
Vamos a crear un dataframe con los festivos que hubo en la ciudad de Madrid entre enero de 2023 y octubre de 2024

```python
# Vamos a crear un dataset con los días de vacaciones
import holidays

# Crear una lista de días festivos en Madrid entre 2023 y 2024
espana_holidays = holidays.Spain(years=[2023, 2024], observed=True, prov='MD')  # 'MD' es la provincia de Madrid

# Filtrar solo los días festivos en el rango de fechas
festivos_madrid = [fecha for fecha, nombre in espana_holidays.items() if '2023-01-01' <= str(fecha) <= '2024-11-30']

# Crear un DataFrame con la columna 'fecha' para los días festivos
df_festivos_madrid = pd.DataFrame(festivos_madrid, columns=['fecha'])

# Mostrar el DataFrame con los días festivos
df_festivos_madrid
```

```python
df_festivos_madrid.to_csv('festivos.csv', index=False, sep=';')
```
