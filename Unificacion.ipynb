```python
# Importamos la librería necesaria
import pandas as pd
import numpy as np
```

### Lectura de los datos

```python
# Leemos los csv con los datos
datos = pd.read_csv('datos_finales.csv', sep=';', decimal='.')
datos_meteo83 = pd.read_csv('datos_meteo_83.csv', sep=';', decimal='.')
datos_meteo89 = pd.read_csv('datos_meteo_89.csv', sep=';', decimal='.')
festivos = pd.read_csv('festivos.csv', sep=';', decimal='.')
```

### Unificación de los datos

```python
# Mostramos los datos leídos de la temperatura
datos_meteo83.head()
```

```python
# Mostramos los datos leídos
datos.head()
```

```python
# Añadimos la temperatura al conjunto de datos
datos_unidos = pd.merge(datos, datos_meteo83[['ANO', 'MES', 'DIA', 'HORA', 'VALOR']], 
                        on=['ANO', 'MES', 'DIA', 'HORA'], how='left')

# Renombrar la columna VALOR a TEMPERATURA en el DataFrame resultante
datos_unidos['TEMPERATURA'] = datos_unidos['VALOR']

# Eliminar la columna VALOR si ya no la necesitas
datos_unidos = datos_unidos.drop(columns=['VALOR'])

# Mostramos los datos
datos_unidos.head()
```

```python
# Mostramos los datos leídos de las precipitaciones
datos_meteo89.head()
```

```python
# Añadimos la humedad al conjunto de datos
datos_unidos = pd.merge(datos_unidos, datos_meteo89[['ANO', 'MES', 'DIA', 'HORA', 'VALOR']], 
                        on=['ANO', 'MES', 'DIA', 'HORA'], how='left')

# Renombrar la columna VALOR a PRECIPITACION en el DataFrame resultante
datos_unidos['PRECIPITACION'] = datos_unidos['VALOR']

# Eliminar la columna VALOR si ya no la necesitas
datos_unidos = datos_unidos.drop(columns=['VALOR'])

datos_unidos.head()
```

```python
# Mostramos los datos leídos de los festivos
festivos.head()
```

```python
# Convertimos el formato de la columna fecha
festivos['fecha'] = pd.to_datetime(festivos['fecha'])

# Dividimos la columna fecha en ANO, MES, DIA
festivos['ANO'] = festivos['fecha'].dt.year
festivos['MES'] = festivos['fecha'].dt.month
festivos['DIA'] = festivos['fecha'].dt.day

# Creamos una columna llamada VALOR que será 1 para todos los días festivos
festivos['VALOR'] = 1
```

```python
# Mostramos los datos
festivos.head()
```

```python
# Unimos los festivos
datos_unidos = pd.merge(datos_unidos, festivos[['ANO', 'MES', 'DIA', 'VALOR']], 
                        on=['ANO', 'MES', 'DIA'], how='left')

# Renombrar la columna VALOR a FESTIVO en el DataFrame resultante
datos_unidos['FESTIVO'] = datos_unidos['VALOR']

# Eliminar la columna VALOR si ya no la necesitas
datos_unidos = datos_unidos.drop(columns=['VALOR'])

# Los valores nulos de la columna FESTIVOS han de tener valor 0
datos_unidos['FESTIVO'] = datos_unidos['FESTIVO'].fillna('0')

# Mostramos los datos
datos_unidos.head()
```

### Tratamiento de datos  
Se realizan una serie de transformaciones una vez se han unificado los datos

```python
# Eliminamos las columnas de periodo de integración y tipo de elemento ya que nos las vamos a utilizar en este trabajo
datos_procesados = datos_unidos.drop(columns=['tipo_elem', 'periodo_integracion'])
```

```python
# Comprobamos los valores nulos
# Contar valores nulos por columna
nulos_por_columna = datos_procesados.isnull().sum()

# Mostrar el resultado
print("Valores nulos por columna:")
print(nulos_por_columna)
```

Rellenamos los valores nulos siguiendo el siguiente criterio:  
- Los datos de precipitación se rellenan a 0
- El resto de datos se rellenan con la media para igual valor de HORA, MINUTO, DIA_SEMANA, FESTIVO y PRECIPITACION

```python
# Rellenamos con cero los valores nulos de la columna PRECIPITACION
datos_procesados['PRECIPITACION'] = datos_procesados['PRECIPITACION'].fillna(0)

# Rellenamos el resto de columnas nulas con la media para igual valor en las columnas HORA, MINUTO, DIA_SEMANA, FESTIVO y PRECIPITACION
datos_procesados['intensidad'] = datos_procesados.groupby(
    ['HORA', 'MINUTO', 'DIA_SEMANA', 'FESTIVO', 'PRECIPITACION']
)['intensidad'].transform(lambda x: x.fillna(x.mean()))

datos_procesados['ocupacion'] = datos_procesados.groupby(
    ['HORA', 'MINUTO', 'DIA_SEMANA', 'FESTIVO', 'PRECIPITACION']
)['ocupacion'].transform(lambda x: x.fillna(x.mean()))

datos_procesados['carga'] = datos_procesados.groupby(
    ['HORA', 'MINUTO', 'DIA_SEMANA', 'FESTIVO', 'PRECIPITACION']
)['carga'].transform(lambda x: x.fillna(x.mean()))

datos_procesados['vmed'] = datos_procesados.groupby(
    ['HORA', 'MINUTO', 'DIA_SEMANA', 'FESTIVO', 'PRECIPITACION']
)['vmed'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Mostramos los datos
datos_procesados.head()
```

Guardamos los datos definitivos

```python
# Pasamos el conjunto de datos final a csv
datos_procesados.to_csv('datos_procesados.csv', index=False, sep=';')
```
